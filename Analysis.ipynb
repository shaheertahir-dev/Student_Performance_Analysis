```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/StudentsPerformance.csv")
print(df.shape)
print(df.head())
print(df.info())
```

TASK2

```python
# ============ TASK 2 - DESCRIPTIVE STATISTICS ============

scores = df[['math score', 'reading score', 'writing score']]

print("=== MEAN ===")
print(scores.mean())

print("\n=== MEDIAN ===")
print(scores.median())

print("\n=== MODE ===")
print(scores.mode().iloc[0])

print("\n=== STD DEVIATION ===")
print(scores.std())

print("\n=== MIN ===")
print(scores.min())

print("\n=== MAX ===")
print(scores.max())
```

```python
# Mean vs Median Comparison
print("Mean vs Median Comparison:")
comparison = pd.DataFrame({
    'Mean': scores.mean(),
    'Median': scores.median(),
    'Difference': scores.mean() - scores.median()
})
print(comparison)
```

```python
# GroupBy - Average scores by Gender
print("Average scores by Gender:")
print(df.groupby('gender')[['math score', 'reading score', 'writing score']].mean().round(2))

# GroupBy - Average scores by Parental Education
print("\nAverage scores by Parental Level of Education:")
print(df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean().round(2))
```

## Task 2 - Observations

1. **Math scores have the highest variability** – The standard deviation of math 
scores (15.16) is the highest among all three subjects, meaning student 
performance in math is more spread out compared to reading and writing.

2. **Reading and writing scores are close to each other** – The mean reading 
score (69.16) and writing score (68.05) are very similar, suggesting students 
who perform well in reading also tend to perform well in writing.

3. **Math mean and median are almost equal (66.08 vs 66.0)** – This tells us 
math scores are fairly normally distributed with no extreme skewness.

4. **Female students outperform male students in reading and writing** – GroupBy 
results show females score higher in reading and writing, while males score 
slightly higher in math.

5. **Higher parental education leads to better scores** – Students whose parents 
have a master's degree or bachelor's degree consistently score higher across 
all three subjects compared to students whose parents only completed high school.

```python
# ============ TASK 3 - DATA CLEANING ============

# Step 1: Create Messy Dataset
import numpy as np

df_messy = df.copy()
np.random.seed(42)

# Add missing values
for col in ['math score', 'reading score', 'writing score']:
    idx = np.random.choice(df_messy.index, size=30, replace=False)
    df_messy.loc[idx, col] = np.nan

# Add duplicate rows
df_messy = pd.concat([df_messy, df_messy.iloc[:20]], ignore_index=True)

# Add outliers
df_messy.loc[0, 'math score'] = 200
df_messy.loc[1, 'reading score'] = -10

# Save messy file
df_messy.to_csv('data/messy_students.csv', index=False)
print("Messy dataset created and saved!")
print("Shape:", df_messy.shape)
```

```python
# Step 2: Show all problems in messy data
df_m = pd.read_csv('data/messy_students.csv')

print("=== SHAPE ===")
print(df_m.shape)

print("\n=== MISSING VALUES ===")
print(df_m.isnull().sum())

print("\n=== DUPLICATES ===")
print("Duplicate rows:", df_m.duplicated().sum())

print("\n=== OUTLIER CHECK (describe) ===")
print(df_m[['math score', 'reading score', 'writing score']].describe())
```

```python
# Step 3: Clean the data

# Remove duplicates
df_clean = df_m.drop_duplicates()
print("After removing duplicates:", df_clean.shape)

# Fix outliers (clip scores to 0-100)
for col in ['math score', 'reading score', 'writing score']:
    df_clean[col] = df_clean[col].clip(0, 100)
print("Outliers fixed!")

# Fill missing values with median
for col in ['math score', 'reading score', 'writing score']:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())
print("Missing values filled!")

# Save cleaned file
df_clean.to_csv('data/cleaned_students.csv', index=False)
print("Cleaned file saved!")
```

```python
# Step 4: Before vs After Summary
print("===== BEFORE CLEANING =====")
print("Shape:", df_m.shape)
print("Missing values:\n", df_m.isnull().sum())
print("Duplicates:", df_m.duplicated().sum())

print("\n===== AFTER CLEANING =====")
print("Shape:", df_clean.shape)
print("Missing values:\n", df_clean.isnull().sum())
print("Duplicates:", df_clean.duplicated().sum())
```

```python
# ============ TASK 4 - EDA & CORRELATION ============

# Load cleaned dataset
df_c = pd.read_csv('data/cleaned_students.csv')
score_cols = ['math score', 'reading score', 'writing score']

# GroupBy Gender
print("=== Average Scores by Gender ===")
print(df_c.groupby('gender')[score_cols].mean().round(2))
```

```python
# GroupBy Race/Ethnicity
print("=== Average Scores by Race/Ethnicity ===")
print(df_c.groupby('race/ethnicity')[score_cols].mean().round(2))
```

```python
# GroupBy Parental Education
print("=== Average Scores by Parental Education ===")
print(df_c.groupby('parental level of education')[score_cols].mean().round(2))
```

```python
# GroupBy Test Preparation Course
print("=== Average Scores by Test Preparation Course ===")
print(df_c.groupby('test preparation course')[score_cols].mean().round(2))
```

```python
# Correlation Matrix
corr_matrix = df_c[score_cols].corr()
print("=== Correlation Matrix ===")
print(corr_matrix.round(3))

# Seaborn Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',
            fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap – Student Scores')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=150)
plt.show()
```

## Task 4 - EDA Insights

1. **Reading and writing scores are highly correlated (0.95)** – Students who 
score well in reading almost always score well in writing too, suggesting 
these two skills develop together.

2. **Math has a weaker correlation with reading/writing (~0.80)** – Math is 
a more independent skill and does not always follow reading/writing performance.

3. **Test preparation course makes a significant difference** – Students who 
completed the test preparation course scored 5-8 points higher on average 
across all three subjects compared to those who did not.

4. **Group E consistently scores the highest** – Among all race/ethnicity 
groups, Group E has the highest average scores in math, reading, and writing.

5. **Parental education directly impacts student performance** – Students 
whose parents hold a master's degree score significantly higher than students 
whose parents only completed high school or some high school.

6. **Female students outperform in reading and writing** – Males score higher 
in math while females consistently lead in reading and writing across all groups.

7. **Lunch type reflects socioeconomic impact** – Students with standard lunch 
score 10+ points higher on average than students with free/reduced lunch, 
suggesting socioeconomic background plays a major role in academic performance.
